```python
from transformers import pipeline

# Load the pre-trained model for summarization
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def generate_meaning(sentence):
    # Use the summarizer to generate the meaning/summary of the sentence
    summary = summarizer(sentence, max_length=50, min_length=10, do_sample=False)
    return summary[0]['summary_text']

# Example sentence
sentence = "Machine learning is a method of data analysis that automates analytical model building."

# Generate meaning
meaning = generate_meaning(sentence)
print("Original Sentence: ", sentence)
print("Generated Meaning: ", meaning)
```
